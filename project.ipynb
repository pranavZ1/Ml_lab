```python
import pandas as pd
import numpy as np
```

```python
eng_abs_df = pd.read_csv('English_Abstractive.csv')
```

```python
eng_abs_df
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import chi2
```

```python
X = eng_abs_df.drop('Judgement Status',axis=1)
```

```python
X
```

```python
Y = eng_abs_df['Judgement Status']
```

```python
Y
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_linear = LinearRegression()
```

```python
sfs = SequentialFeatureSelector(model_linear,n_features_to_select=100,direction='forward')
```

```python
sfs.fit(X,Y)
```

```python
selected_features = X.columns[sfs.get_support()]
```

```python
selected_features
```

```python
eng_abs_transformed_df = eng_abs_df[selected_features]
```

```python
eng_abs_transformed_df
```

```python

X_transformed = eng_abs_transformed_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train_transformed,X_test_transformed,Y_train_transformed,Y_test_transformed = train_test_split(X_transformed,Y,test_size=0.3,random_state=42)
```

```python
Y_test
```

```python
# KNN Classifier for the above problem on the original dataset
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix, roc_curve, auc
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,Y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
accuracy_knn = accuracy_score(Y_test,y_pred_knn)
```

```python
print("The accuracy score for a KNN Classifier with k=5 on the original dataset is : ",accuracy_knn*100,"%")
```

```python
# To check for the optimal k value
k_values = range(1,300)
accuracies_knn = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,Y_train)
    y_pred_optimal_knn = knn.predict(X_test)
    accuracy_optimal_knn = accuracy_score(Y_test,y_pred_optimal_knn)
    accuracies_knn.append(accuracy_optimal_knn)
```

```python
#visualization
import matplotlib.pyplot as plt
```

```python
plt.plot(k_values,accuracies_knn)
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title("Accuracy vs k")
```

```python
# The optimal k value can be attained at k=114
knn_optimal = KNeighborsClassifier(n_neighbors=114)
```

```python
knn_optimal.fit(X_train,Y_train)
```

```python
y_pred_knn_optimal = knn_optimal.predict(X_test)
```

```python
accuracy_optimal_114_knn = accuracy_score(Y_test,y_pred_knn_optimal)
```

```python
print("The maximum accuracy score for a KNN Classifier with optimal k=114 on the original dataset is : ",accuracy_optimal_114_knn*100,"%")
```

```python
# KNN Classifier for on the transformed dataset
knn_transformed = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_transformed.fit(X_train_transformed,Y_train_transformed)
```

```python
y_pred_knn_transformed = knn_transformed.predict(X_test_transformed)
```

```python
accuracy_knn_transformed = accuracy_score(Y_test_transformed,y_pred_knn_transformed)
```

```python
accuracy_knn_transformed
```

```python
# To check for the optimal k value for the transformed dataset
k_values = range(1,300)
accuracies_knn_transformed = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_transformed,Y_train_transformed)
    y_pred_optimal_knn_transformed = knn.predict(X_test_transformed)
    accuracy_optimal_knn_transformed = accuracy_score(Y_test,y_pred_optimal_knn_transformed)
    accuracies_knn_transformed.append(accuracy_optimal_knn_transformed)
```

```python
plt.plot(k_values,accuracies_knn_transformed)
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title("Accuracy vs k")
```

```python
plt.plot(k_values,accuracies_knn_transformed,label = 'transformed accuracies')
plt.plot(k_values,accuracies_knn,label = 'accuracies')
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy , Transformed_accuracy")
plt.title("Accuracy vs Transformed Accuracy vs k")
plt.legend()
```
