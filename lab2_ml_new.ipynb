```python
data = df4.iloc[:20,8:].values
data
```

```python
num_vec = data.shape[0]
num_vec
```

```python
jc_matrix = np.zeros((num_vec, num_vec))
smc_matrix = np.zeros((num_vec, num_vec))
cosine_matrix = np.zeros((num_vec, num_vec))

for i in range(num_vec):
    for j in range(i, num_vec):
        vec1 = data[i]
        vec2 = data[j]

        intersection = np.sum(vec1 & vec2)
        union = np.sum(vec1 | vec2)
        jc_matrix[i, j] = jc_matrix[j, i] = intersection / union if union != 0 else 0

        matching = np.sum(vec1 == vec2)
        smc_matrix[i, j] = smc_matrix[j, i] = matching / len(vec1)

        dot_product = np.dot(vec1, vec2)
        magnitude_vec1 = np.sqrt(np.sum(vec1 ** 2))
        magnitude_vec2 = np.sqrt(np.sum(vec2 ** 2))
        cosine_matrix[i, j] = cosine_matrix[j, i] = dot_product / (magnitude_vec1 * magnitude_vec2)

```
