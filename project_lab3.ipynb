```python
c1_vecc = np.array(df_class1['0'])
c2_vecc = np.array(df_class1['1'])
```

```python
c1_vecc
```

```python
c2_vecc
```

```python
def min_dis(a,b,r):
    return np.sum(np.abs(a-b)**r)**(1/r)

r_values = np.arange(1,11)
distances = [min_dis(c1_vecc,c2_vecc,r) for r in r_values]

plt.figure(figsize=(8,8))
plt.plot(r_values,distances,marker='o')
plt.xlabel('r')
plt.ylabel('Min distance')
plt.title('Min distance between 2 feature vector')
plt.xticks(r_values)
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df1.iloc[:, :-1].values
y = df1['Judgement Status'].values 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
accuracy = knn.score(X_test, y_test)
print(f"Accuracy of kNN (k=3): {accuracy * 100:.2f}%")
```

```python
y_pred = knn.predict(X_test)
```

```python
for i in range(10):
    print(f"Predicted: {y_pred[i]}, Actual: {y_test[i]}")
```

```python
accuracies = []
k_values = range(1, 12)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracies.append(knn.score(X_test, y_test))

# Plotting the accuracy
plt.figure(figsize=(8, 8))
plt.plot(k_values, accuracies, marker='o')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy of kNN Classifier with varying k')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```
